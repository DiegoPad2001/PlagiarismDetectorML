```python
import csv
import os


def reemplazar_ids_por_codigos(archivo_csv):
    with open(archivo_csv, 'r') as archivo:
        lector = csv.reader(archivo)
        filas = list(lector)
        for fila in filas[1:]:
            # Reemplazar en la primera columna
            id_archivo = fila[0]
            id_archivo_2 = fila[1]
            ruta_archivo_codigo = os.path.join(
                'version_1/'+ id_archivo +'_'+id_archivo_2+'/'+ id_archivo + '.java')
            try:
                with open(ruta_archivo_codigo, encoding="utf8") as archivo_codigo:
                    codigo = archivo_codigo.read()
                    fila[0] = codigo
            except FileNotFoundError:
                fila[0] = "no encontrado"

            # Reemplazar en la segunda columna

            ruta_archivo_codigo_2 = os.path.join(
                'version_1/' + id_archivo + '_'+id_archivo_2+'/' + id_archivo_2 + '.java')
            try:
                with open(ruta_archivo_codigo_2, encoding="utf8") as archivo_codigo_2:
                    codigo_2 = archivo_codigo_2.read()
                    fila[1] = codigo_2
            except FileNotFoundError:
                fila[1] = "no encontrado"

    nuevo_archivo_csv = 'nuevo_' + archivo_csv


    with open(nuevo_archivo_csv, 'w', newline='', encoding='utf8') as archivo_salida:
        escritor = csv.writer(archivo_salida)
        escritor.writerows(filas)

    print(f"Se ha creado el nuevo archivo CSV: {nuevo_archivo_csv}")


nombre_archivo_csv = 'labels.csv'
reemplazar_ids_por_codigos(nombre_archivo_csv)
```
