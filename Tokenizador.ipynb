```python
import pandas as pd
import javalang as javalang

data = pd.read_csv('nuevo_labels.csv')
tree = javalang.parse.parse("package javalang.brewtab.com; class Test {}")
```

```python
import re

# Expresiones regulares para los tokens
token_patterns = [
    (r'\/\/.*', 'COMMENT'),  # Comentarios de una línea
    (r'\/\*(.|\n)*?\*\/', 'COMMENT'),  # Comentarios de varias líneas
    (r'\b(abstract|assert|boolean|break|byte|case|catch|char|class|const|continue|default|do|double|else|enum|extends|final|finally|float|for|if|implements|import|instanceof|int|interface|long|native|new|package|private|protected|public|return|short|static|strictfp|super|switch|synchronized|this|throw|throws|transient|try|void|volatile|while)\b', 'KEYWORD'),  # Palabras reservadas
    (r'([a-zA-Z_][a-zA-Z0-9_]*)(?=\s*\()', 'FUNCTION'),  # Funciones
    (r'\+|-|\*|/|%|\+=|-=|\*=|/=|%=|==|!=|>|<|>=|<=', 'OPERATOR'),  # Operaciones
    (r'\b[a-zA-Z_][a-zA-Z0-9_]*\b', 'VARIABLE'),  # Variables
    (r'\b\d+\b', 'NUMBER'),  # Números
    (r'\".*?\"', 'STRING'),  # Strings
]

# Función para tokenizar el código Java
def tokenize_java_code(code):
    tokens = []
    position = 0

    while position < len(code):
        match = None
        token_type = None

        for pattern, token_name in token_patterns:
            regex = re.compile(pattern)
            match = regex.match(code, position)
            if match:
                token_type = token_name
                value = match.group(0)
                tokens.append((token_type, value))
                break

        if match:
            position = match.end()
        else:
            position += 1
        
    aux = ""
    for token in tokens:
        if token[0] == 'KEYWORD':
            aux += token[1].upper() + " "
        else:
            aux += token[0] + " "
    return aux

    #return tokens

# Ejemplo de uso
#code = data["sub1"][0]

#tokens = tokenize_java_code(code)

# Imprimir los tokens
#for token in tokens:
   # print(token[0])

    

data['sub1'] = data['sub1'].apply(tokenize_java_code)
data['sub2'] = data['sub2'].apply(tokenize_java_code)
```
