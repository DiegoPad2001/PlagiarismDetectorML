```python
import pandas as pd
import javalang as javalang


data = pd.read_csv('nuevo_labels.csv')
```

```python

# Función para tokenizar el código Java
def tokenize_java_code(java_code):
    tokens = []
    try:
        token_generator = javalang.tokenizer.tokenize(java_code)
        for token in token_generator:
            tokens.append(token.__class__.__name__)
    except Exception as e:
        print("Tokenization error:", e)
    return tokens

    #return tokens

# Ejemplo de uso
#code = data["sub1"][0]

#tokens = tokenize_java_code(code)

# Imprimir los tokens
#for token in tokens:
   # print(token[0])

    

data['sub1'] = data['sub1'].apply(tokenize_java_code)
data['sub2'] = data['sub2'].apply(tokenize_java_code)
```
