```python
import tkinter as tk
from tkinter import ttk
import pickle
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
import xgboost as xgb

# Cargar el modelo entrenado
with open('modelo.pkl', 'rb') as f:
    clf = pickle.load(f)

# Cargar el vectorizador
with open('vectorizer.pkl', 'rb') as f:
    vectorizer = pickle.load(f)


def compare_codes():
    code1 = text_box1.get("1.0", tk.END).strip()
    code2 = text_box2.get("1.0", tk.END).strip()

    if code1 and code2:
        new_data = pd.DataFrame({'sub1': [code1], 'sub2': [code2]})

        # Vectorize the new data
        new_data_vectorized = vectorizer.transform(
            new_data['sub1'] + ' ' + new_data['sub2'])

        # Make predictions
        predictions = clf.predict(new_data_vectorized)

        if predictions[0] == 1:
            result_text.set("Comparación:\n Es plagio")
        else:
            result_text.set("Comparación:\n No es plagio")

    else:
        result_text.set("¡Ingresa dos códigos para comparar!")


# Crear la interfaz
window = tk.Tk()
window.title("Comparador de Códigos")
window.config(bg='black')

# Estilo personalizado para el botón
style = ttk.Style()
style.theme_use('clam')
style.configure('Custom.TButton', font=('Arial', 14), background='#007ACC', relief='raised', borderwidth=0, foreground='white')

# Cuadro de texto y etiqueta para el código 1
label1 = tk.Label(window, text="Código 1:", bg='black', fg='white', font=('Arial', 14))
label1.grid(row=0, column=0, sticky='w')
text_box1 = tk.Text(window, height=10, width=50, bg='black', fg='white', font=('Arial', 12))
text_box1.grid(row=1, column=0, padx=10, pady=5, sticky='nsew')

# Cuadro de texto y etiqueta para el código 2
label2 = tk.Label(window, text="Código 2:", bg='black', fg='white', font=('Arial', 14))
label2.grid(row=0, column=1, sticky='w')
text_box2 = tk.Text(window, height=10, width=50, bg='black', fg='white', font=('Arial', 12))
text_box2.grid(row=1, column=1, padx=10, pady=5, sticky='nsew')

# Botón para comparar los códigos
compare_button = ttk.Button(window, text="Comparar", command=compare_codes, style='Custom.TButton')
compare_button.grid(row=2, column=0, columnspan=2, pady=10, sticky='ew')

# Resultado de la comparación
result_text = tk.StringVar()
result_label = tk.Label(window, textvariable=result_text, bg='black', fg='white', font=('Arial', 14))
result_label.grid(row=3, column=0, columnspan=2)

# Ajustar el tamaño y espaciado de las columnas y filas
window.grid_columnconfigure(0, weight=1)
window.grid_columnconfigure(1, weight=1)
window.grid_rowconfigure(1, weight=1)

# Ajustar el tamaño de la ventana
window.geometry("600x400")

# Iniciar la interfaz
window.mainloop()
```
